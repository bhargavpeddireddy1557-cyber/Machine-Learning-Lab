```python
import pandas as pd
```

```python
from sklearn.datasets import fetch_openml
```

```python
from sklearn.model_selection import train_test_split
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
import warnings
```

```python
warnings.filterwarnings("ignore")
```

```python
boston_raw = fetch_openml(name="boston", version=1, as_frame=True)
```

```python
X_boston = boston_raw.data
```

```python
y_boston = boston_raw.target.astype(float)
```

```python
scaler_boston = StandardScaler()
```

```python
X_boston_scaled = scaler_boston.fit_transform(X_boston)
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    X_boston_scaled,
    y_boston,
    test_size=0.20,
    random_state=42
)
```

```python
print("Training set shape:", X_train.shape)
```

```python
print("Test set shape:", X_test.shape)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
```

```python
lr_model = LinearRegression()
```

```python
lr_model.fit(X_train, y_train)
```

```python
lr_preds = lr_model.predict(X_test)
```

```python
mse = mean_squared_error(y_test, lr_preds)
r2 = r2_score(y_test, lr_preds)
```

```python
print(f"Mean Squared Error (MSE): {mse:.2f}")
print(f"R2 Score: {r2:.2f}")
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import warnings
warnings.filterwarnings("ignore")

# Load Boston Housing Dataset from GitHub
url = "https://raw.githubusercontent.com/selva86/datasets/master/BostonHousing.csv"
df = pd.read_csv(url)

# Features and Target
X = df.drop("medv", axis=1)
y = df["medv"]

# Standardize Features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Train-Test Split (80% Training, 20% Testing)
X_train, X_test, y_train, y_test = train_test_split(
    X_scaled,
    y,
    test_size=0.20,
    random_state=42
)

print("Training set shape:", X_train.shape)
print("Test set shape:", X_test.shape)
```
